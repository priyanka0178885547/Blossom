```python
from textblob import TextBlob

def get_sentiment(text):
    score = TextBlob(text).sentiment.polarity

    if score > 0:
        return "Positive"
    elif score == 0:
        return "Neutral"
    else:
        return "Negative"
```

```python
import numpy as np

def extract_features(reviews):
    cleaned = [clean_text(r) for r in reviews]

    X = vectorizer.transform(cleaned)
    feature_names = vectorizer.get_feature_names_out()

    scores = np.mean(X.toarray(), axis=0)
    top_idx = scores.argsort()[-8:]

    return [feature_names[i] for i in top_idx]
```

```python
def feature_sentiment(reviews, features):
    summary = {}

    for f in features:
        sentiments = []

        for r in reviews:
            if f in r.lower():
                sentiments.append(get_sentiment(r))

        if sentiments:
            pos = sentiments.count("Positive")
            neg = sentiments.count("Negative")

            if pos > neg:
                summary[f] = "Positive"
            elif neg > pos:
                summary[f] = "Negative"
            else:
                summary[f] = "Mixed"

    return summary
```

```python
def generate_explanation(feature_summary):
    explanation = []

    for f, s in feature_summary.items():
        if s == "Positive":
            explanation.append(f"{f} is well appreciated by users.")
        elif s == "Negative":
            explanation.append(f"{f} has many complaints.")
        else:
            explanation.append(f"{f} shows mixed feedback.")

    return explanation
```

```python
def analyze_reviews(reviews):
    results = []
    predictions = []

    for r in reviews:
        label, conf, _ = predict_fake(r)
        sentiment = get_sentiment(r)

        results.append({
            "review": r,
            "label": label,
            "confidence": conf,
            "sentiment": sentiment
        })

        predictions.append(1 if "Fake" in label else 0)

    return results, predictions
```

```python
def compute_trust(predictions):
    fake = sum(predictions)
    total = len(predictions)

    fake_percent = (fake / total) * 100
    trust = 100 - fake_percent

    return round(trust,2), round(fake_percent,2)
```

```python
def run_pipeline(reviews):

    results, preds = analyze_reviews(reviews)

    trust, fake_percent = compute_trust(preds)

    features = extract_features(reviews)
    feature_summary = feature_sentiment(reviews, features)

    explanation = generate_explanation(feature_summary)

    return {
        "Trust Score": trust,
        "Fake %": fake_percent,
        "Top Features": features[:5],
        "Explainable AI": explanation,
        "Details": results
    }
```
