```python
import pandas as pd

df = pd.read_csv("/content/Preprocessed Fake Reviews Detection Dataset.csv")
print(df.head())
print(df.columns)
```

```python
print(df.columns)
```

```python
df = df[['text_', 'label', 'rating']]
df.columns = ['review', 'fake', 'rating']
```

```python
df = df.sample(min(20000, len(df)))
print(df.shape)
```

```python
df = df[['review', 'fake', 'rating']]
df.dropna(inplace=True)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

df['cleaned'] = df['review'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)

X = vectorizer.fit_transform(df['cleaned'])
y = df['fake']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

```python
def sentiment_label(rating):
    if rating >= 4:
        return "Positive"
    elif rating == 3:
        return "Neutral"
    else:
        return "Negative"

df['sentiment'] = df['rating'].apply(sentiment_label)
```

```python
fake_count = df['fake'].sum()
total = len(df)

trust = 100 - (fake_count / total * 100)
print("Trust Score:", trust)
```

```python
feature_names = vectorizer.get_feature_names_out()
coefficients = model.coef_[0]

# Top words indicating fake reviews
top_fake_words = sorted(zip(coefficients, feature_names), reverse=True)[:10]

print("Top Fake Indicators:")
for coef, word in top_fake_words:
    print(word)
```

```python
def explain_prediction(text):
    words = text.split()
    important = []

    for w in words:
        if w in feature_names[:100]:   # important words
            important.append(w)

    return important
```

```python
sample = "This product is amazing and perfect"

print("Explanation:", explain_prediction(sample))
```
